### Image analysis
now we will analyse the images sorted to see every genre styles and features like colors and shapes

```python
import cv2
import numpy as np
from collections import defaultdict
```

```python
base_dir = r'C:\Users\Anderson\Desktop\archive (3)\organized'
```

```python
def calculate_color_histogram(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Calculate histogram for each color channel
    hist_r = cv2.calcHist([image], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([image], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([image], [2], None, [256], [0, 256])
    
    # Normalize histograms and flatten them
    hist_r /= hist_r.sum()
    hist_g /= hist_g.sum()
    hist_b /= hist_b.sum()
    
    return np.concatenate([hist_r.flatten(), hist_g.flatten(), hist_b.flatten()])
```

```python
def detect_shapes(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    edges = cv2.Canny(image, 100, 200)
    
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Extract contour information (e.g., number of contours, average area)
    num_contours = len(contours)
    if num_contours > 0:
        avg_area = np.mean([cv2.contourArea(cnt) for cnt in contours])
    else:
        avg_area = 0
    
    return num_contours, avg_area
```

```python
# Initialize lists to hold feature data and labels
features = []
labels = []
```

```python
# Iterate over each genre folder
for genre in os.listdir(base_dir):
    genre_dir = os.path.join(base_dir, genre)
    
    if os.path.isdir(genre_dir):  # Check if it is a directory
        for image_filename in os.listdir(genre_dir):
            image_path = os.path.join(genre_dir, image_filename)
            
            # Extract features
            color_histogram = calculate_color_histogram(image_path)
            num_contours, avg_contour_area = detect_shapes(image_path)
            
            # Combine features into a single array
            feature_vector = np.concatenate([color_histogram, [num_contours, avg_contour_area]])
            
            # Append the feature vector and the corresponding genre label
            features.append(feature_vector)
            labels.append(genre)
```

```python
# Convert the lists to a DataFrame (optional)
df_features = pd.DataFrame(features)
df_labels = pd.Series(labels, name='Genre')
```

```python
# Combining features and labels into a single DataFrame
df_combined = pd.concat([df_features, df_labels], axis=1)
```

```python
# Saving the DataFrame to a CSV file for future use
df_combined.to_csv('art_style_features.csv', index=False)
```

### Data Augmentation

```python
import os
import cv2
from imgaug import augmenters as iaa
from tqdm import tqdm
```

```python
# Function to augment images in all genre directories
def augment_images_in_all_genres(base_dir, num_augmented=5, min_images=100):
    """
    Applies augmentation to images in each genre subfolder within the base directory.
    Augmented images are saved in the same directory.

    Args:
    - base_dir: Path to the base directory containing genre subfolders.
    - num_augmented: Number of augmented images to generate for each original image.
    - min_images: Minimum number of images required in a folder. Folders with fewer images will be augmented.
    """
    seq = iaa.Sequential([
        iaa.Fliplr(0.5),
        iaa.Crop(percent=(0, 0.1)),
        iaa.Affine(rotate=(-20, 20)),
        iaa.GaussianBlur(sigma=(0.0, 3.0))
    ])

    for genre in os.listdir(base_dir):
        genre_dir = os.path.join(base_dir, genre)
        
        if os.path.isdir(genre_dir):
            image_files = [img for img in os.listdir(genre_dir) if img.endswith(('.jpg', '.jpeg', '.png'))]
            
            if len(image_files) < min_images:
                print(f"Augmenting images in {genre}...")
                for img_name in tqdm(image_files):
                    img_path = os.path.join(genre_dir, img_name)
                    image = cv2.imread(img_path)
                    
                    for i in range(num_augmented):
                        images_aug = seq(images=[image])
                        augmented_img_name = f"{os.path.splitext(img_name)[0]}_aug_{i}.jpg"
                        cv2.imwrite(os.path.join(genre_dir, augmented_img_name), images_aug[0])
```

```python
# Example usage: Apply augmentation to all underrepresented genres
base_dir = r'C:\Users\Anderson\Desktop\archive (3)\organized'
augment_images_in_all_genres(base_dir, num_augmented=5, min_images=100)
```

### Streamlined CNN Example with TensorFlow/Keras

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Path to your image folder
image_folder = r'C:\Users\Anderson\Desktop\archive (3)\organized'
```

```python
# Define image size and batch size
image_size = (128, 128)  # Adjust image size as needed
batch_size = 32
```

```python
# Create an ImageDataGenerator with validation split
datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    validation_split=0.2,  # Reserve 20% of the data for validation
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Load training images
train_generator = datagen.flow_from_directory(
    image_folder,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  # Use this data for training
)
```

```python
# Load validation images
validation_generator = datagen.flow_from_directory(
    image_folder,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  # Use this data for validation
)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    epochs=10,  # Adjust epochs as needed
    validation_data=validation_generator
)
```

```python
# Evaluate the model on the validation data
val_loss, val_acc = model.evaluate(validation_generator, verbose=2)
print(f'Validation accuracy: {val_acc}')
```

```python
import requests
from io import BytesIO
from PIL import Image
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
# Function to load and preprocess the image from a URL
def load_image_from_url(url, target_size=(128, 128)):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    img = img.resize(target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Normalize the image array
    return img_array, img  # Return both the preprocessed array and the original image
```

```python
# Function to predict the art style
def predict_art_style(model, img_array):
    prediction = model.predict(img_array)
    predicted_class_index = np.argmax(prediction, axis=1)[0]
    class_labels = train_generator.class_indices
    class_labels = {v: k for k, v in class_labels.items()}  # Invert the dictionary
    predicted_class_label = class_labels[predicted_class_index]
    return predicted_class_label
```
